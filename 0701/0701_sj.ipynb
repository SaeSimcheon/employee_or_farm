# 짝지어 제거하기


```python
from collections import deque

def solution(s):
    S = []
    s = deque(s)
    while s:
        a = s.popleft()
        if S:
            if S[-1] == a:
                S.pop()
            else:
                S.append(a)
        else:
            S.append(a)
    return 0 if S else 1

# deque가 가장 먼저 생각났음
# deque에 넣고 맞으면 빼고 안맞으면 다시 넣고 반복

# 좋아요 가장 많은 풀이
def solution(s):
    answer = []
    for i in s:
        if not(answer):
            answer.append(i)
        else:
            if(answer[-1] == i):
                answer.pop()
            else:
                answer.append(i)
    return not(answer)

# que를 안쓰고도 푸는게 가능하네 근데 사람 생각은 비슷하나봄봄
```

# 타겟 넘버


```python
def solution(numbers, target):
    def DFS(L, res):
        nonlocal cnt
        if L == len(numbers):
            if res == target:
                cnt += 1
        else:
            DFS(L+1, res+numbers[L])
            DFS(L+1, res-numbers[L])
    cnt = 0
    DFS(0, 0)
    return cnt

# 배운점
# 1. DFS 할때 함수안에 넣으면 global이 아니라 nonlocal로 전역함수를 설정해야함
# 그 이외에는 일반 DFS 풀이

# 좋아요 가장 많은 풀이
def solution(numbers, target):
    if not numbers and target == 0 :
        return 1
    elif not numbers:
        return 0
    else:
        return solution(numbers[1:], target-numbers[0]) + solution(numbers[1:], target+numbers[0])

# WOW
# 이런 식으로도 풀수가 있네;

answer = 0
def DFS(idx, numbers, target, value):
    global answer
    N = len(numbers)
    if(idx== N and target == value):
        answer += 1
        return
    if(idx == N):
        return

    DFS(idx+1,numbers,target,value+numbers[idx])
    DFS(idx+1,numbers,target,value-numbers[idx])
def solution(numbers, target):
    global answer
    DFS(0,numbers,target,0)
    return answer

# nonlocal을 안쓸땐 solution 함수에도 global 처리를 해야함
```

```python
import heapq as hq
def solution(scoville, K):
    answer = 0
    hq.heapify(scoville)
    while True:
        minimum = hq.heappop(scoville)
        if minimum >= K:
            return answer
        elif len(scoville) == 0 and minimum < K:
            return -1
        else:
            sec_minimum = hq.heappop(scoville)
            new_scoville = minimum + sec_minimum*2
            hq.heappush(scoville,new_scoville)
            answer += 1

# heap을 제대로 써본적이 없었는데 공부하고 써봤다
# que로도 해봤는데, 도저히 시간을 맞출수가 없던데 heap이 이런 이진탐색에서는 중요하긴 한가보다

# 좋아요 가장 많은 풀이
import heapq as hq

def solution(scoville, K):

    hq.heapify(scoville)
    answer = 0
    while True:
        first = hq.heappop(scoville)
        if first >= K:
            break
        if len(scoville) == 0:
            return -1
        second = hq.heappop(scoville)
        hq.heappush(scoville, first + second*2)
        answer += 1

    return answer
# 전반적으로 heap을 쓰면 풀이가 다 비슷비슷한듯
```

# 게임 맵 최단거리


```python
from collections import deque
def solution(maps):
    dx = [0,1,0,-1]
    dy = [1,0,-1,0]
    m = len(maps)
    n = len(maps[0])
    distance = [[1]*n for _ in range(m)]
    Q = deque()
    Q.append((0,0))
    maps[0][0] = 0
    while Q:
        current = Q.popleft()
        for i in range(4):
            x = current[0] + dx[i]
            y = current[1] + dy[i]
            if 0 <= x <= m-1 and 0 <= y <= n-1 and maps[x][y]==1:
                maps[x][y] = 0
                distance[x][y] = distance[current[0]][current[1]] + 1
                Q.append((x,y))

    if distance[m-1][n-1] == 1:
        return -1
    else:
        return distance[m-1][n-1]

# 예전에 풀었던 BFS 문제랑 동일 미로 최단거리 문제인가
# 파이썬은 이중리스트 길이가 헷갈린다

# 다른 사람 풀이
from collections import deque

def solution(maps):
    x_move = [1, 0, -1, 0]
    y_move = [0, 1, 0, -1]

    x_h, y_h = (len(maps[0]), len(maps))
    queue = deque([(0, 0, 1)])

    while queue:
        x, y, d = queue.popleft()

        for i in range(4):
            nx = x + x_move[i]
            ny = y + y_move[i]

            if nx > -1 and ny > -1 and nx < x_h and ny < y_h:
                if maps[ny][nx] == 1 or maps[ny][nx] > d + 1:
                    maps[ny][nx] = d + 1
                    if nx == x_h - 1 and ny == y_h - 1:
                        return d + 1

                    queue.append((nx, ny, d + 1))

    return -1

# 비슷한 풀이
```

# [1차] 뉴스 클러스터링


```python
from collections import Counter
def solution(str1, str2):
    str1_ = []
    str2_ = []
    for i in range(len(str1)-1):
        component = str1[i:i+2].lower()
        if component.isalpha():
            str1_.append(component)
    for i in range(len(str2)-1):
        component = str2[i:i+2].lower()
        if component.isalpha():
            str2_.append(component)
    inter = list((Counter(str1_)&Counter(str2_)).elements())
    union = list((Counter(str1_)|Counter(str2_)).elements())
    if union: # 교집합은 합집합의 부분집합이니까 합집합이 공집합이 되는경우만 주의
        return int((len(inter)/len(union))*65536)
    else:
        return 65536

# 정규표현식으로 풀고싶었는데 헷갈려서 그냥 isalpha로 썼음
# Counter output이 dictionary인데, .elements()를 쓰면 key들을 원소로 받을수있더라

# 좋아요 가장 많은 풀이

import re
import math

def solution(str1, str2):
    str1 = [str1[i:i+2].lower() for i in range(0, len(str1)-1) if not re.findall('[^a-zA-Z]+', str1[i:i+2])]
    str2 = [str2[i:i+2].lower() for i in range(0, len(str2)-1) if not re.findall('[^a-zA-Z]+', str2[i:i+2])]

    gyo = set(str1) & set(str2)
    hap = set(str1) | set(str2)

    if len(hap) == 0 :
        return 65536

    gyo_sum = sum([min(str1.count(gg), str2.count(gg)) for gg in gyo])
    hap_sum = sum([max(str1.count(hh), str2.count(hh)) for hh in hap])

    return math.floor((gyo_sum/hap_sum)*65536)

# 고수의 냄새가 물씬 풍긴다
# 처음에 구상했던 풀이... 아직 많이 후달린다는걸 느낀다

# 반성점
# 1. 정규표현식에 좀더 익숙해져야겠다

# 배운점
# 1. count.elements()
# # 개수를 카운트 했던 Counter객체의 dictionary 자료를 다시 반복시켜서 리스트 반환
```

# 행렬 테두리 회전하기


```python
def solution(rows, columns, queries):
    answer = []
    matrix = [list(range(i*columns+1, (i+1)*columns+1)) for i in range(rows)]
    for x1, y1, x2, y2 in queries:
        start = matrix[x1-1][y1-1]
        tmp = start
        for i in range(y1,y2):
            matrix[x1-1][i], start = start, matrix[x1-1][i]
            if tmp > start:
                tmp = start
        for i in range(x1,x2):
            matrix[i][y2-1], start = start, matrix[i][y2-1]
            if tmp > start:
                tmp = start
        for i in range(y2-2, y1-2, -1):
            matrix[x2-1][i], start = start, matrix[x2-1][i]
            if tmp > start:
                tmp = start
        for i in range(x2-2, x1-2, -1):
            matrix[i][y1-1], start = start, matrix[i][y1-1]
            if tmp > start:
                tmp = start
        answer.append(tmp)
    return answer

# 문제 지시사항대로 위 -> 오른쪽 -> 아래 -> 왼쪽 순으로 회전해서 값을 바꾸는...

# 좋아요 가장 많은 풀이
def solution(rows, columns, queries):
    answer = []
    board = [[i+(j)*columns for i in range(1,columns+1)] for j in range(rows)]
    # print(board)
    for a,b,c,d in queries:
        stack = []
        r1, c1, r2, c2 = a-1, b-1, c-1, d-1
        for i in range(c1, c2+1):
            stack.append(board[r1][i])
            if len(stack) == 1:
                continue
            else:
                board[r1][i] = stack[-2]
        for j in range(r1+1, r2+1):
            stack.append(board[j][i])
            board[j][i] = stack[-2]
        for k in range(c2-1, c1-1, -1):
            stack.append(board[j][k])
            board[j][k] = stack[-2]
        for l in range(r2-1, r1-1, -1):
            stack.append(board[l][k])
            board[l][k] = stack[-2]
        answer.append(min(stack))
    return answer

# 비슷한데 스택으로 해결했구만
```

# 괄호 변환


```python
def solution(p):
    def split(column):
        l, r = 0, 0
        for i in column:
            if i == "(":
                l += 1
            else:
                r += 1
            if l == r:
                break
        u, v = column[:l+r], column[l+r:]

        return (u, v)
    def check(u_column):
        col = u_column
        stack = []
        idx = 0
        while idx < len(col):
            start = col[idx]
            if len(stack) == 0:
                stack.append(start)
            elif stack[-1]=="(" and start == ")":
                stack.pop(-1)
            else:
                stack.append(start)
            idx += 1
        if stack:
            return False
        else:
            return True
    def convert(u_column):
        col = list(u_column)
        col.pop(0)
        col.pop(-1)
        if not col:
            return ''
        else:
            for i in range(len(col)):
                if col[i] == "(":
                    col[i] = ")"
                else:
                    col[i] = "("
            return ''.join(col)

    def correct(col):
        if col == '':
            return ''
        else:
            u, v = split(col)
            if check(u):
                return u + correct(v)
            else:
                new_u = convert(u)
                return '(' + correct(v) + ')' + new_u
    return correct(p)

# 문제 이해를 바로 못해서 질문하기 보고 풀었음
# 이해에 문제가 있는건지 문제에 문제가 있는건지

def solution(p):
    if p=='': return p
    r=True; c=0
    for i in range(len(p)):
        if p[i]=='(': c-=1
        else: c+=1
        if c>0: r=False
        if c==0:
            if r:
                return p[:i+1]+solution(p[i+1:])
            else:
                return '('+solution(p[i+1:])+')'+''.join(list(map(lambda x:'(' if x==')' else ')',p[1:i]) ))

# ? 이해 못함
```

# 메뉴 리뉴얼

```python
from itertools import combinations as cb
from collections import Counter
def solution(orders, course):
    orders = ["".join(sorted(i)) for i in orders]
    max_order = max(map(lambda x: len(x),orders))
    answer = []
    for i in course:
        if i > max_order:
            break
        cb_list = []
        for j in orders:
            if len(j) >= i:
                cb_list += list(cb(j,i))
        counter = Counter(cb_list).most_common()
        max_count = counter[0][1]
        for c in counter:
            if max_count > 1 and c[1] == max_count:
                answer.append("".join(c[0]))
    return sorted(answer)

# 문제를 보자마자 떠오른건 모든 경우의수(조합)을 다 고려해야만 하는구나...
# 다행히 길이가 적어서 복잡도는 그리 크지 않은듯
# 근데 존나 무지성으로 풀긴했다 이리저리 복잡한 문제여 ㅅㅂ

import collections
import itertools

def solution(orders, course):
    result = []

    for course_size in course:
        order_combinations = []
        for order in orders:
            order_combinations += itertools.combinations(sorted(order), course_size)

        most_ordered = collections.Counter(order_combinations).most_common()
        result += [ k for k, v in most_ordered if v > 1 and v == most_ordered[0][1] ]

    return [ ''.join(v) for v in sorted(result) ]

# 비슷한거 같은데 오히려 짧다...
# 왜 그럴까 생각해봤는데 처음에 sort를 굳이 안해도 되는거 같네 ( 어차피 counter에서 경우의수 다 추려주니까 )
```
