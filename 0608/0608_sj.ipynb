# 소수 만들기

```python
from itertools import combinations as cb
def isPrime(x):
    if x==1:
        return False
    for i in range(2, x):
        if x%i==0:
            return False
    return True

def solution(nums):
    comb = list(cb(nums, 3))
    answer = 0
    for i in comb:
        if isPrime(sum(i)):
            answer += 1
    return answer

# 문제 딱 보고 생각난게 DFS로 갈거냐 조합으로 할거냐 였는데
# 그냥 간단하게 조합으로 풀기로 했음
# 다른 답안도 비슷한 풀이로 풀었더라
```

# 키패드 누르기

```python
def solution(numbers, hand):
    answer = ''
    left = [0,-1]
    right = [0,1]
    pad = [[0,0],[3,-1],[3,0],[3,1],[2,-1],[2,0],[2,1],[1,-1],[1,0],[1,1]]
    for i in numbers:
        if i in [1,4,7]:
            answer += 'L'
            left = pad[i]
        elif i in [3,6,9]:
            answer += 'R'
            right = pad[i]
        else:
            xleft = abs(left[0]-pad[i][0])+abs(left[1]-pad[i][1])
            xright = abs(right[0]-pad[i][0])+abs(right[1]-pad[i][1])
            if xleft < xright:
                answer += 'L'
                left = pad[i]
            elif xleft > xright:
                answer += 'R'
                right = pad[i]
            else:
                if hand == 'left':
                    answer += 'L'
                    left = pad[i]
                else:
                    answer += 'R'
                    right = pad[i]
    return answer

# 개어렵다
# 처음에 거리계산을 어떻게하지 고민하다가 좌표로 생각하기로 했음
# 숫자 0을 (0,0)으로 정의하면 패드간 거리는 dx+dy가 되니까 OK
# 중간중간에 왼손 오른손도 꼭 업데이트하기
# 초기 시작은 *랑 # 좌표로 기록해놔야 index out of range가 안뜸

# 좋아요 가장 많은 풀이
def solution(numbers, hand):
    answer = ''
    key_dict = {1:(0,0),2:(0,1),3:(0,2),
                4:(1,0),5:(1,1),6:(1,2),
                7:(2,0),8:(2,1),9:(2,2),
                '*':(3,0),0:(3,1),'#':(3,2)}

    left = [1,4,7]
    right = [3,6,9]
    lhand = '*'
    rhand = '#'
    for i in numbers:
        if i in left:
            answer += 'L'
            lhand = i
        elif i in right:
            answer += 'R'
            rhand = i
        else:
            curPos = key_dict[i]
            lPos = key_dict[lhand]
            rPos = key_dict[rhand]
            ldist = abs(curPos[0]-lPos[0]) + abs(curPos[1]-lPos[1])
            rdist = abs(curPos[0]-rPos[0]) + abs(curPos[1]-rPos[1])

            if ldist < rdist:
                answer += 'L'
                lhand = i
            elif ldist > rdist:
                answer += 'R'
                rhand = i
            else:
                if hand == 'left':
                    answer += 'L'
                    lhand = i
                else:
                    answer += 'R'
                    rhand = i

    return answer

# 딕셔너리로 저장하는걸 생각했는데 그냥 이중리스트가 쉬워보여서 안했음
# 생각해보니 이렇게 키를 저장하면 진짜 버튼 누르는거 같다
# 대부분 답안이 비슷하게 접근하는거 같음
```

# 실패율

```python
from collections import Counter
def solution(N, stages):
    answer = []
    failure = dict()
    D = Counter(stages)
    a_len = len(stages)
    for i in range(N):
        if a_len == 0:
            failure[i+1] = 0
            continue
        failure[i+1] = D[i+1]/a_len
        a_len -= D[i+1]
    failure_ = sorted(failure.items(), key = lambda x:x[1], reverse = True)
    answer = [failure_[i][0] for i in range(N)]
    return answer

# 처음에 70점 맞고 나머지가 런타임 에러나서 뭔가 싶었는데
# 아무도 도달못한 stage가 있으면 0/0이라서 런타임 에러가 나는거 같다

# 좋아요 가장 많은 풀이

def solution(N, stages):
    result = {}
    denominator = len(stages)
    for stage in range(1, N+1):
        if denominator != 0:
            count = stages.count(stage)
            result[stage] = count / denominator
            denominator -= count
        else:
            result[stage] = 0
    return sorted(result, key=lambda x : result[x], reverse=True)

# 직관적인 풀이
def solution(N, stages):
    fail = {}
    for i in range(1,N+1):
        try:
            fail_ = len([a for a in stages if a==i])/len([a for a in stages if a>=i])
        except:
            fail_ = 0
        fail[i]=fail_
    answer = sorted(fail, key=fail.get, reverse=True)
    return answer
# 리스트로 직접 실패율을 계산하는 방법 / try, except를 사용해서 보다 직관적이다

# 반성점
# 1. 언제나 문제의 예외경우를 생각해야 한다(이번 문제에서는 도달못한 stage의 경우)

# 배운점
# 1. result는 dictionary이므로 sorted에 result를 그냥 넘기면 result의 keys가 들어갑니다. keys는 생략이 가능합니다.
# 거기에 lambda는 기준을 result[x]: 즉 value로 정렬한다는 뜻입니다. 그래서 key가 출력되게 됩니다.
```

# [1차] 비밀지도


```python
def ten_to_bit(n,k):
    arr = [0]*n
    while k>0:
        a = k%2
        k = k//2
        arr[n-1] = a
        n -= 1
    return arr
def solution(n, arr1, arr2):
    answer = []
    for i in range(n):
        ar1 = ten_to_bit(n, arr1[i])
        ar2 = ten_to_bit(n, arr2[i])
        answer_ = ''
        for j in range(n):
            if ar1[j]==0 and ar2[j]==0:
                answer_ += ' '
            else:
                answer_ += '#'
        answer.append(answer_)
    return answer

# 이진수 변환을 직접 함수로 만들어서 짰다

# 좋아요 가장 많은 풀이
def solution(n, arr1, arr2):
    answer = []
    for i,j in zip(arr1,arr2):
        a12 = str(bin(i|j)[2:])
        a12=a12.rjust(n,'0')
        a12=a12.replace('1','#')
        a12=a12.replace('0',' ')
        answer.append(a12)
    return answer

# 내장함수를 적절히 사용한다면 이렇게도 짤수가 있구나... 오늘도 배워갑니다

# 배운점
# 1. bin(number) : 전달받은 integer를 이진수 문자열로 돌려줌.
# ex) bin(10) -> '0b1010' 그래서 bin(i)[2:] 이렇게 한다
# bin(i|j) : 두수의 이진수를 구한뒤 or 연산을 함께,,,,
# 2. rjust, ljust : 오른쪽 정렬, 왼쪽 정렬
# ex) a12.rjust(n,'0') -> 문자열 a12를 크기가 n인 문자열로 만들고
# 오른쪽 정렬시 생성되는 공간을 '0'으로 채운다
# 3. zip() : iterable 여러개를 한번에 접근가능하게 해주는 함수
# 반복문에서 두개 이상의 리스트를 (i,j,...) 이렇게 접근 가능
```
