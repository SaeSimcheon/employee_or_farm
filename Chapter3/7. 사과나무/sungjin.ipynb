```python
%%timeit
import sys
sys.stdin = open("in1.txt", "rt")
n = int(sys.stdin.readline())
a = [list(map(int, sys.stdin.readline().split())) for i in range(n)]
apple = 0
median = int((n-1)/2)
for i in range(median+1):
    if i < median:
        apple += sum(a[i][(median-i):(median+1+i)])
        apple += sum(a[n-i-1][(median-i):(median+1+i)])
    else :
        apple += sum(a[i][(median-i):(median+1+i)])
print(apple)

# median에서 각 극단이 list index가 같다는 점에서 접근
# median일때는 하나만 더해주면 되겠지?
```
