```python
%%timeit

import sys
sys.stdin = open("in1.txt", "rt")
a = [list(map(int, sys.stdin.readline().split())) for i in range(9)]
b = list(range(1,10))
def row(x): 
    if all(sorted(i) == b for i in x):
        return True
def col(x):
    cnt = 0
    for i in range(9):
        if sorted(list(a[j][i] for j in range(9))) == b:
            cnt += 1
        else:
            cnt += 0
    if cnt == 9:
        return True
    else:
        return False
def square(x):
    cnt = 0
    n = m = 3
    for i in range(0, 9, n):
        for j in range(0, 9, m):
            c = sum([row[j:j + m] for row in a[i:i + n]], [])
            if sorted(c) == b:
                cnt += 1
            else:
                cnt += 0
    if cnt == 9:
        return True
    else:
        return False

if row(a) and col(a) and square(a):
    print("YES")
else:
    print("NO")


```
