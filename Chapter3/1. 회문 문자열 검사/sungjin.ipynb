```python
%%timeit
import sys

sys.stdin = open("in1.txt","rt")
n = int(sys.stdin.readline())
for i in range(n):
    a = sys.stdin.readline()
    a = list(a.lower())
    if a == a[::-1]:
        print("#%d YES" %(i+1))
    else:
        print("#%d NO" %(i+1))
```

첫 시도는 100 / 회문은 [::-1] 하면 간단
근데 시간 체크 이렇게 하는거 맞나?
