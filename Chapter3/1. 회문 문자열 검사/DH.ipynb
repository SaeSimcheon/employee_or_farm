## 회문 문자열

```python
import sys

for num in range(1,6):
    sys.stdin = open("in{}.txt".format(num),"rt")

    N = int(sys.stdin.readline())
    ans = open("out{}.txt".format(num),"rt") 

    def find(x):
        while 1:
            if len(x) == 1 or len(x) == 0:
                return "YES"

            s = x.pop(0)
            e = x.pop(-1)

            if s != e:
                return "NO"

    for i in range(N):

        if i == N-1:

            L = list(sys.stdin.readline().lower())

            RE = find(L)

            print("#{0} {1}".format(i+1,RE), ans.readline() , sep = "     ")
            print()
        else: 

            L = list(sys.stdin.readline().lower())[:-1]

            RE = find(L)

            print("#{0} {1}".format(i+1,RE), ans.readline() , sep = "     ")
```

정답

간단 s == s[::-1]

```python
import sys
sys.stdin = open("in1.txt","rt")

n = int(input())

for i in range(n):
    s = input()
    s = s.upper()
    
    size = len(s)
    
    for j in range(size//2):
        if s[j] != s[-1-j]:
            print("#%d NO" %(i+1))
    else :
        print("#%d YES" %(i+1))
```
