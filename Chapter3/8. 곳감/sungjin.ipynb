```python
%%timeit
import sys
sys.stdin = open("in1.txt", "rt")
n = int(sys.stdin.readline())
a = [list(map(int, sys.stdin.readline().split())) for i in range(n)]
m = int(sys.stdin.readline())
b = [list(map(int, sys.stdin.readline().split())) for i in range(m)]
for i in b:
    if n < i[2]:
        i[2] = i[2]%n # 크면 한바퀴 돌고 나머지로만 도니까
    if i[1] == 0:
        a[i[0]-1] = a[i[0]-1][i[2]:] + a[i[0]-1][:i[2]]
    else:
        a[i[0]-1] = a[i[0]-1][(n-i[2]):] + a[i[0]-1][:(n-i[2])]
persimmon = 0
median = int((n-1)/2)
for i in range(median+1):
    if i < median:
        persimmon += sum(a[i][i:(n-i)])
        persimmon += sum(a[n-i-1][i:(n-i)])
    else :
        persimmon += sum(a[i][i:(i+1)])
print(persimmon)
```
