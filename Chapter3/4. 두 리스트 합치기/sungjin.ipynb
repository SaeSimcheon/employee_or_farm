```python
%%timeit
import sys
sys.stdin = open("in1.txt", "rt")
n = int(sys.stdin.readline())
a = list(map(int, sys.stdin.readline().split()))
m = int(sys.stdin.readline())
b = list(map(int, sys.stdin.readline().split()))
a.extend(b)
new = sorted(a)
for i in new:
    print(i, end=' ')
```
