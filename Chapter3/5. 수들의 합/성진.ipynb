```python
%%timeit
import sys
sys.stdin = open("in1.txt", "rt")
n, m = map(int, sys.stdin.readline().split())
a = list(map(int, sys.stdin.readline().split()))

i = 0
j = 1
summ = a[i]
cnt = 0

while True:
    if summ < m:
        if j < n:
            summ += a[j]
            j += 1
        else:
            break
    elif summ == m:
        cnt += 1
        summ -= a[i]
        i += 1
    else:
        summ -= a[i]
        i += 1
print(cnt)
```
