```python
%%timeit
import sys
sys.stdin = open("in1.txt", "rt")
n = int(sys.stdin.readline())
a = [list(map(int, sys.stdin.readline().split())) for i in range(n)]
f = [[0]*(n+2) for _ in range(n+2)]
for i in range(len(f)-2):
    f[i+1] = [0]+a[i]+[0]

cnt = 0
for i in range(n):
    for j in range(n):
        if f[i+1][j+1] > max(f[i+2][j+1],f[i+1][j+2],f[i][j+1],f[i+1][j]):
            cnt += 1
print(cnt)
```

2) (n+2)*(n+2)의 영행렬 만들고 거기에 원래 행렬 집어넣으면 상하좌우 값만 비교하면 된다
