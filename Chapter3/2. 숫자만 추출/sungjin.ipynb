```python
%%timeit
import sys

sys.stdin = open("in1.txt", "rt")
a = sys.stdin.readline()
b = [i for i in a if i.isdigit()]
natural = int(''.join(b))
print(natural)
cnt = 0
for i in range(1,natural+1):
    if natural%i == 0:
        cnt+=1
print(cnt)
```

첫 시도 100 
1) .isdigit -> 숫자인지 아닌지 체크, string으로 되어있어도 체크 가능
2) ''.join -> 예전에 본거 같은데, 리스트 합치기
