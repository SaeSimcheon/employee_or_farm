```python
%%timeit
import sys

sys.stdin = open("in1.txt", "rt")
a = sys.stdin.readline()
b = [i for i in a if i.isdigit()]
natural = int(''.join(b))
print(natural)
cnt = 0
for i in range(1,natural+1):
    if natural%i == 0:
        cnt+=1
print(cnt)
```
