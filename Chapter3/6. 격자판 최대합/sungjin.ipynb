```python
%%timeit
import sys
sys.stdin = open("in1.txt", "rt")
n = int(sys.stdin.readline())
a = [list(map(int, sys.stdin.readline().split())) for i in range(n)]

allsum = []
for i in range(n):
    rowsum = 0
    colsum = 0
    for j in range(n):
        rowsum += a[i][j]
        colsum += a[j][i]
    allsum.append(rowsum)
    allsum.append(colsum)
diagsum1 = 0
diagsum2 = 0
for i in range(n):
    diagsum1 += a[i][i]
    diagsum2 += a[i][n-1-i]
allsum.append(diagsum1)
allsum.append(diagsum2)
print(max(allsum))
```
