```python
%%timeit
import sys
sys.stdin = open("in1.txt", "rt")
a = [list(map(int, sys.stdin.readline().split())) for i in range(7)]
c = []
for i in range(7):
    c.append(list(a[j][i] for j in range(7))) # transpose
def count(x):
    cnt = 0
    for i in x:
        for j in range(3):
            b = i[j:j+5]
            if b == b[::-1]:
                cnt += 1
            else:
                cnt += 0
    return cnt

print(count(a)+count(c))
```
