```python
class Solution
{
    public int solution(String s)
    {
        if (s.length() == 1){
            return 1;
        }
        if (s.length() == 2){
            if (s.substring(0,1) == s.substring(1,2)){
                return 2;
            }else
            {
                return 1;
            }
            
        }
        int answer = 1;
        
        //System.out.println(s.substring(0,2));
        
        int M1 ;
        int M2 ;
        int can1 ;
        int can2 ;
        int cancan;
        for ( int i = 1 ; i < s.length()-1; i ++ ){
            can1 =1 ;
            can2 =1 ;
            cancan = 0 ; 
            M1 = Math.min(s.length()- i-1,i );
            M2 = Math.min(s.length()- i ,i );
            if (Math.max(M1,M2) < answer){
                break;
            }
            for (int j = 1 ; j < M1+1; j++){
                
                if ((i-j < 0)||(i+j+1 > s.length())){continue;}
                if (s.substring(i-j, i-j+1).equals(s.substring(i+j,i+j+1))){
                    can1 =can1 +2 ;
                }else{
                    break;}
            }
            for (int k = 1 ; k < M2+1; k++){
                if ((i-k < 0)||(i+k+2 > s.length())){continue;}
                if (s.substring(i-k,i).equals(s.substring(i+k,i+k+2))){
                    can2 =can2 +2 ;
                }else{
                    break;}
            }
            
            cancan = Math.max(can1,can2);
            //System.out.println(i+" "+s.substring(i,i+1)+" "+cancan);
            if (answer < cancan){
                answer= cancan;
            }
            
        }
        //System.out.println(answer);
        
        return answer;
    }
}
```

```python
import java.util.*;

class Solution {
    int[][] dp;
        public int solution(int n, int[] money) {
            dp = new int[money.length+1][n+1];
            int answer = 0;
            for(int i = 1; i < money.length+1; i++){
                for(int j = 0; j < n+1; j++){
                    if(j == 0){
                        dp[i][j] = 1;
                    }
                    else {
                        if(j < money[i-1]){
                            dp[i][j] = dp[i-1][j];
                        }
                            
                        else{
                            dp[i][j] = (dp[i-1][j] + dp[i][j-money[i-1]]) % 1000000007;
                        }
                            
                    }
                }
            }
            return dp[money.length][n];
        }
}
```
