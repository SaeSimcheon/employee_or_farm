```python
def solution(a):
    if len(a)==1:
        return 1
    answer = 1
    M=a.index(min(a))
    
    L=a[:M]
    R=a[M+1:][::-1]
    if len(L) !=0:    
        L_st = L[0]
        answer +=1
        for idx,i in enumerate(L):
            if idx == 0:
                continue
            if L_st < i:
                continue
            else :
                L_st = i
                answer +=1
    if len(R)!=0:
        R_st = R[0]
        answer +=1
        for idx,i in enumerate(R):
            if idx == 0:
                continue
            if R_st < i:
                continue
            else:
                R_st =i
                answer+=1
    return answer
```
