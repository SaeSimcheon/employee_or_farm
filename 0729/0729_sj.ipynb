# [1차] 추석 트래픽


```python
def check(time, times):
    cnt = 0
    start, end = time, time+1000
    for t in times:
        if t[0] < end and t[1] >= start:
            cnt += 1
    return cnt
def solution(lines):
    answer = 0
    times = []
    for line in lines:
        __, ee, tt = line.split(" ")
        ee = ee.split(":")
        ee = (int(ee[0])*60*60 + int(ee[1])*60 + float(ee[2]))*1000
        tt = float(tt[:-1])*1000
        ss = ee-tt+1
        times.append([int(ss),int(ee)])
    sorted(times, key = lambda x: (x[1],x[0]))
    for time in times:
        answer = max(answer, check(time[0],times), check(time[1],times))
    return answer

# 문자열 파싱하고 비교하는거 까지는 쉽게 생각했는데(예전 그리디 문제 떠올렸음)
# 시작시간 - 종료시간만 비교하다가 계속 틀려서 둘다 비교하는걸로 했더니 맞더라

# 좋아요 가장 많은 풀이
def solution(lines):
    #get input
    S , E= [], []
    totalLines = 0
    for line in lines:
        totalLines += 1
        type(line)
        (d,s,t) = line.split(" ")
       ##time to float
        t = float(t[0:-1])
        (hh, mm, ss) = s.split(":")
        seconds = float(hh) * 3600 + float(mm) * 60 + float(ss)
        E.append(seconds + 1)
        S.append(seconds - t + 0.001)
    #count the maxTraffic
    S.sort()
    curTraffic = 0
    maxTraffic = 0
    countE = 0
    countS = 0
    while((countE < totalLines) & (countS < totalLines)):
        if(S[countS] < E[countE]):
            curTraffic += 1
            maxTraffic = max(maxTraffic, curTraffic)
            countS += 1
        else: ## it means that a line is over.
            curTraffic -= 1
            countE += 1
    return maxTraffic

# 탐색이 비슷하게 간거 같은데도 되게 신박해보인다
```

# 가장 먼 노드

```python
from collections import deque
def solution(n, edge):
    answer = [1]+[0]*(n-1)
    matrix = [[] for _ in range(n)]
    for x, y in edge:
        matrix[x-1].append(y-1)
        matrix[y-1].append(x-1)
    Q = deque([0])
    while Q:
        length = len(Q)
        for i in range(length):
            q = Q.popleft()
            for j in matrix[q]:
                if answer[j] == 0:
                    answer[j] = 1
                    Q.append(j)
    return length

# 처음에 플로이드 와샬을 해봤는데, 자연스레 시간초과 + 다틀림
# BFS로 거리 1, 2,... 하는식으로 가장 먼 거리를 찾았음

# 좋아요 가장 많은 풀이
def solution(n, edge):
    graph =[  [] for _ in range(n + 1) ]
    distances = [ 0 for _ in range(n) ]
    is_visit = [False for _ in range(n)]
    queue = [0]
    is_visit[0] = True
    for (a, b) in edge:
        graph[a-1].append(b-1)
        graph[b-1].append(a-1)
    while queue:
        i = queue.pop(0)
        for j in graph[i]:
            if is_visit[j] == False:
                is_visit[j] = True
                queue.append(j)
                distances[j] = distances[i] + 1
    distances.sort(reverse=True)
    answer = distances.count(distances[0])
    return answer

# 꽤 비슷한 풀이인듯?

from collections import defaultdict
def bfs(graph, start, distances):
    q = [start]
    visited = set([start])
    while len(q) > 0:
        current = q.pop(0)
        for neighbor in graph[current]:
            if neighbor not in visited:
                visited.add(neighbor)
                q.append(neighbor)
                distances[neighbor] = distances[current] + 1

def solution(n, edge):
    # 그래프 만들기
    graph = defaultdict(list)
    for e in edge:
        graph[e[0]].append(e[1])
        graph[e[1]].append(e[0])
    # bfs 탐색 (최단 거리를 구해야 하므로.)
    distances = [0]*(n+1)
    bfs(graph, 1, distances)
    max_distance = max(distances)
    answer = 0
    for distance in distances:
        if distance == max_distance:
            answer += 1
    return answer
# 깔끔하다 보고 배우자
```

# 입국심사

```python
def solution(n, times):
    answer = max(times)*n
    start = 1
    end = max(times)*n
    while start <= end:
        mid = (start+end)//2
        cnt = sum(mid//time for time in times)
        if n <= cnt:
            answer = min(answer,mid)
            end = mid-1
        else:
            start = mid+1
    return answer

# 감을 못잡아서 질문하기에 힌트보고 풀었음
# 초기값 설정을 이리저리 해보다가 가장 오래걸리는 심사관으로 했더니 되더라.. 이해가 잘안감

# 좋아요 가장 많은 풀이
def solution(n, times):
    answer = 0
    start, end, mid = 1, times[-1] * n, 0
    while start < end:
        mid = (start + end) // 2
        total = 0
        for time in times:
            total += mid // time
        if total >= n:
            end = mid
        else:
            start = mid + 1
    answer = start
    return answer
```

```python
from collections import deque
def solution(n, computers):
    answer = 0
    board = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            if computers[i][j] == 1:
                board[i][j] = 1
                board[j][i] = 1
    Q = deque()
    for i in range(n):
        if board[i][i] == 1:
            board[i][i] = 0
            Q.append(i)
            while Q:
                x = Q.popleft()
                for y in range(n):
                    if board[x][y] == 1:
                        board[x][y] = 0
                        board[y][y] = 0
                        board[y][x] = 0
                        Q.append(y)
            answer += 1
    return answer

# 일반 좌표 행렬이 아닌, 인접 행렬 형태
# 질문하기 보다보니 대각행렬이 아닐수도 있다고 해서 일부러 대각행렬 만들어줌

# 좋아요 가장 많은 풀이
def solution(n, computers):
    answer = 0
    visited = [0 for i in range(n)]
    def dfs(computers, visited, start):
        stack = [start]
        while stack:
            j = stack.pop()
            if visited[j] == 0:
                visited[j] = 1
            # for i in range(len(computers)-1, -1, -1):
            for i in range(0, len(computers)):
                if computers[j][i] ==1 and visited[i] == 0:
                    stack.append(i)
    i=0
    while 0 in visited:
        if visited[i] ==0:
            dfs(computers, visited, i)
            answer +=1
        i+=1
    return answer

# 생각보다 비슷한 BFS 풀이인듯? visited에 미리 방문한걸 저장하는 식


def solution(n, computers):
    temp = []
    for i in range(n):
        temp.append(i)
    for i in range(n):
        for j in range(n):
            if computers[i][j]:
                for k in range(n):
                    if temp[k] == temp[i]:
                        temp[k] = temp[j]
    return len(set(temp))
# OMG
# 플루이드-와샬 응용이라는데 어떻게 돌아가는지는 잘 모르겠다
```

# 디스크 컨트롤러

```python
import heapq
def solution(jobs):
    answer = 0
    complete = 0
    i = 0
    start = -1
    ls = []
    while i < len(jobs):
        for job in jobs:
            if start < job[0] <= complete:
                heapq.heappush(ls, [job[1], job[0]]) # 정렬은 처리시간 기준
        if ls:
            now = heapq.heappop(ls)
            start = complete # 처리된 시점에서 다시 시작
            complete += now[0] # 처리시간끼리만 더하기
            answer += complete-now[1] # 완료 시점에서 대기시점 제외
            i += 1
        else:
            complete += 1 # 대기가 없으면 처리시간 +1
    return answer//len(jobs)

# 문제가 정확히 뭔지 이해가 아직도 안됨... 다만 힙정렬로 푸는거라는건 알겠는데...

# 좋아요 가장 많은 풀이
import heapq
from collections import deque

def solution(jobs):
    tasks = deque(sorted([(x[1], x[0]) for x in jobs], key=lambda x: (x[1], x[0])))
    q = []
    heapq.heappush(q, tasks.popleft())
    current_time, total_response_time = 0, 0
    while len(q) > 0:
        dur, arr = heapq.heappop(q)
        current_time = max(current_time + dur, arr + dur)
        total_response_time += current_time - arr
        while len(tasks) > 0 and tasks[0][1] <= current_time:
            heapq.heappush(q, tasks.popleft())
        if len(tasks) > 0 and len(q) == 0:
            heapq.heappush(q, tasks.popleft())
    return total_response_time // len(jobs)
```
