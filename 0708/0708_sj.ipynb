# 행렬 테두리 회전하기

```python
def solution(rows, columns, queries):
    answer = []
    matrix = [list(range(i*columns+1, (i+1)*columns+1)) for i in range(rows)]
    for x1, y1, x2, y2 in queries:
        start = matrix[x1-1][y1-1]
        tmp = start
        for i in range(y1,y2):
            matrix[x1-1][i], start = start, matrix[x1-1][i]
            if tmp > start:
                tmp = start
        for i in range(x1,x2):
            matrix[i][y2-1], start = start, matrix[i][y2-1]
            if tmp > start:
                tmp = start
        for i in range(y2-2, y1-2, -1):
            matrix[x2-1][i], start = start, matrix[x2-1][i]
            if tmp > start:
                tmp = start
        for i in range(x2-2, x1-2, -1):
            matrix[i][y1-1], start = start, matrix[i][y1-1]
            if tmp > start:
                tmp = start
        answer.append(tmp)
    return answer

# 문제 지시사항대로 위 -> 오른쪽 -> 아래 -> 왼쪽 순으로 회전해서 값을 바꾸는...

# 좋아요 가장 많은 풀이
def solution(rows, columns, queries):
    answer = []
    board = [[i+(j)*columns for i in range(1,columns+1)] for j in range(rows)]
    # print(board)
    for a,b,c,d in queries:
        stack = []
        r1, c1, r2, c2 = a-1, b-1, c-1, d-1
        for i in range(c1, c2+1):
            stack.append(board[r1][i])
            if len(stack) == 1:
                continue
            else:
                board[r1][i] = stack[-2]
        for j in range(r1+1, r2+1):
            stack.append(board[j][i])
            board[j][i] = stack[-2]
        for k in range(c2-1, c1-1, -1):
            stack.append(board[j][k])
            board[j][k] = stack[-2]
        for l in range(r2-1, r1-1, -1):
            stack.append(board[l][k])
            board[l][k] = stack[-2]
        answer.append(min(stack))
    return answer

# 비슷한데 스택으로 해결했구만
```

# 다리를 지나는 트럭

```python
from collections import deque
def solution(bridge_length, weight, truck_weights):
    bridge = [0]*bridge_length
    bridge = deque(bridge)
    truck_weights = deque(truck_weights)
    bridge_weights = 0
    bridge.popleft()
    truck = truck_weights.popleft()
    bridge_weights += truck
    bridge.append(truck)
    answer = 1
    while bridge_weights > 0:
        out = bridge.popleft()
        bridge_weights -= out
        answer += 1
        if truck_weights:
            if bridge_weights + truck_weights[0] <= weight:
                truck = truck_weights.popleft()
                bridge.append(truck)
                bridge_weights += truck
            else:
                bridge.append(0)
    return answer

# 문제 설명 한번보고 알아들으면 머리가 비상한 사람일거 같다
# 난 이해가 안되서 코드짜기도 전에 질문하기 보고 했음

# 좋아요 가장 많은 풀이
import collections
DUMMY_TRUCK = 0
class Bridge(object):
    def __init__(self, length, weight):
        self._max_length = length
        self._max_weight = weight
        self._queue = collections.deque()
        self._current_weight = 0
    def push(self, truck):
        next_weight = self._current_weight + truck
        if next_weight <= self._max_weight and len(self._queue) < self._max_length:
            self._queue.append(truck)
            self._current_weight = next_weight
            return True
        else:
            return False
    def pop(self):
        item = self._queue.popleft()
        self._current_weight -= item
        return item
    def __len__(self):
        return len(self._queue)
    def __repr__(self):
        return 'Bridge({}/{} : [{}])'.format(self._current_weight, self._max_weight, list(self._queue))
def solution(bridge_length, weight, truck_weights):
    bridge = Bridge(bridge_length, weight)
    trucks = collections.deque(w for w in truck_weights)
    for _ in range(bridge_length):
        bridge.push(DUMMY_TRUCK)
    count = 0
    while trucks:
        bridge.pop()
        if bridge.push(trucks[0]):
            trucks.popleft()
        else:
            bridge.push(DUMMY_TRUCK)
        count += 1
    while bridge:
        bridge.pop()
        count += 1
    return count

# 클래스 써서 하는 사람은 처음본다
```

```python
def erase_board(m, n, board):
    ch = [[0]*n for _ in range(m)]
    for i in range(m-1):
        for j in range(n-1):
            if board[i][j] == board[i+1][j] == board[i][j+1] == board[i+1][j+1] != '0':
                ch[i][j] = 1
                ch[i+1][j] = 1
                ch[i][j+1] = 1
                ch[i+1][j+1] = 1
    for i in range(m):
        for j in range(n):
            if ch[i][j] == 1:
                board[i][j] = '0'
    return board


def new_board(m, n, board):
    for j in range(n):
        ch = []
        for i in range(m-1,-1,-1):
            if not ch:
                if board[i][j] == '0':
                    ch.append([i,j])
            else:
                if board[i][j] == '0':
                    ch.append([i,j])
                else:
                    x, y = ch.pop(0)
                    board[x][y] = board[i][j]
                    board[i][j] = '0'
                    ch.append([i,j])
    return board


def solution(m, n, board):
    board = list(map(lambda x: list(x), board))
    for _ in range(n):
        erase_board(m, n, board)
        new_board(m, n, board)
    return sum(map(lambda x: x.count('0'),board))


# 5번이 계속 틀림 ㅜ
```
