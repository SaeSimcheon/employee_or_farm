길찾기게임

```python
# 좌표를 이용하는 문제로 보임
# 정렬을 이용해서 푸는데 
# x,y 좌표를 이용하며 풀어야

# 전위 7 4 6 9 이러면
# 일단 y좌표가 가장 큰 7의 갑 6
# 다음이 4와 2가 y좌표가 큼 / 4가 2보다 x 좌표가 작아서 선택
# 내려가면서 체크한다는 것은 동일 
# 전위때는 현
import sys
sys.setrecursionlimit(10**6)
from collections import deque
def solution(nodeinfo):
    nodeinfo = [[idx+1,i,j] for idx,(i,j) in enumerate(nodeinfo)]
    nodeinfo.sort(key = lambda x : (-x[2],x[1]))
    D1=nodeinfo[:]
    D2=nodeinfo[:]
    D1 = deque(D1)
    D2 = deque(D2)
    
    if len(nodeinfo) == 1:
        return [[1],[1]]

    answer1 = [] 
    
    def dfs1(seq):
        nonlocal answer1
        if len(seq) == 0 :
            return 
        else:
            left = deque([])
            right = deque([])
            answer1.append(seq[0][0])
            pp=list(seq.popleft())
            
            for i in seq:
                if pp[1] < i[1] :
                    right.append(i)
                else:
                    left.append(i)
            
            dfs1(left)
            dfs1(right)    
    answer2 = [] 
    def dfs2(seq):
        nonlocal answer2
        if len(seq) == 0 :
            return 
        else:
            left = deque([])
            right = deque([])
            pp=list(seq.popleft())
            
            for i in seq:
                if pp[1] < i[1] :
                    right.append(i)
                else:
                    left.append(i)
            dfs2(left)
            dfs2(right)    
            answer2.append(pp[0])
    dfs1(D1)
    dfs2(D2)
    answer = [answer1,answer2] 
    return answer
```
