# 순위

```python
def solution(n, results):
    answer = 0
    board = [[0]*n for _ in range(n)]
    for x, y in results:
        board[x-1][y-1] = 1
        board[y-1][x-1] = -1
    for i in range(n):
        for j in range(n):
            for k in range(n):
                if board[i][j] != 0 and board[i][j] == board[j][k]:
                    board[i][k] = board[i][j]
                    board[k][i] = -1*board[i][j]
    for i in board:
        if i.count(0) == 1:
            answer += 1
    return answer

# 그래프 문제 -> 그래프 탐색(BFS)를 할지 플로이드-와샬을 할지??
# 최소경로를 찾는게 아닌 경로마다 체크를 하는식으로
# 근데 로직은 딱 떠올랐는데 코드 구현이 힘들었음

# 좋아요 가장 많은 풀이
from collections import defaultdict
def solution(n, results):
    answer = 0
    win, lose = defaultdict(set), defaultdict(set)
    for result in results:
            lose[result[1]].add(result[0])
            win[result[0]].add(result[1])

    for i in range(1, n + 1):
        for winner in lose[i]: win[winner].update(win[i])
        for loser in win[i]: lose[loser].update(lose[i])

    for i in range(1, n+1):
        if len(win[i]) + len(lose[i]) == n - 1: answer += 1
    return answer

# 대부분 이렇게 해쉬형태로 많이 풀더라
```

# 보석 쇼핑

```python
def solution(gems):
    answer, result = [], []
    D = {}
    wanted = set(gems)
    n, m = len(gems), len(wanted)
    lt, rt = 0, 0
    D[gems[lt]] = 1
    while lt <= rt and rt < n:
        if len(D) < m:
            rt += 1
            if rt >= n:
                break
            if D.get(gems[rt]):
                D[gems[rt]] += 1
            else:
                D[gems[rt]] = 1
        else:
            result.append([lt + 1, rt+1])
            if D[gems[lt]] > 1:
                D[gems[lt]] -= 1
                lt += 1
            else:
                del D[gems[lt]]
                lt += 1
    minimum = 100000
    for i in range(len(result)):
        dis = result[i][1] - result[i][0]
        if dis < minimum:
            minimum = dis
            answer.append(result[i])
    return answer[-1]

# 효율성 맞추기 너무 어렵다;;
# 투포인터 or 슬라이딩 윈도우인데 슬라이싱 할때는 리스트보다 딕셔너리가 더 효율적이었다

# 좋아요 가장 많은 풀이
def solution(gems):
    size = len(set(gems))
    dic = {gems[0]:1}
    temp = [0, len(gems) - 1]
    start , end = 0, 0

    while(start < len(gems) and end < len(gems)):
        if len(dic) == size:
            if end - start < temp[1] - temp[0]:
                temp = [start, end]
            if dic[gems[start]] == 1:
                del dic[gems[start]]
            else:
                dic[gems[start]] -= 1
            start += 1

        else:
            end += 1
            if end == len(gems):
                break
            if gems[end] in dic.keys():
                dic[gems[end]] += 1
            else:
                dic[gems[end]] = 1

    return [temp[0]+1, temp[1]+1]

# 대부분 이렇게 푸는거 같다
# 투포인트 / 슬라이딩 윈도우 참고
# https://blog.naver.com/kks227/220795165570
```

# [1차] 셔틀버스

```python
from collections import deque
def transform(time):
    a, b = divmod(time, 60)
    if a < 10 and b < 10:
        return '0'+str(a)+':'+'0'+str(b)
    elif a >= 10 and b < 10:
        return str(a)+":"+'0'+str(b)
    elif a < 10 and b >= 10:
        return '0'+str(a)+':'+str(b)
    else:
        return str(a)+":"+str(b)

def solution(n, t, m, timetable):
    buses = [540 + t*j for j in range(n)]
    timetable = [int(i.split(":")[0])*60 + int(i.split(":")[1]) for i in timetable]
    timetable.sort()
    timetable = deque(timetable)
    bus = 0
    crew = 0
    numCrew = 0
    for i in range(n):
        bus = buses.pop(0)
        numCrew = 0
        while timetable and numCrew < m:
            if timetable[0] <= bus:
                numCrew += 1
                crew = timetable.popleft()
            else:
                break
    if numCrew < m:
        return transform(bus)
    elif numCrew == m:
        crew -= 1
        return transform(crew)

# 변환 -> 정렬 -> 버스마다 조건 맞으면 태우기해서
# 반복문 끝나면 다 타는지 아닌지로 시간 리턴
# 특별한 풀이는 없는듯?? 그냥 조건 잘 생각해서 코드 짜는게 다인거 같다
```
