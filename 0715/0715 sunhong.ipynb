## 쿼드 압축

```python
## 다 0이면 cnt_0 갯수 +1
## 다 1이면 cnt_1 갯수 +1
## 아니면 다시 DFS하는 식으로

def count_0(arr):
    c_0=0
    for x in arr:
        c_0+=x.count(0)
    return c_0


def DFS(arr): ## 무조건 나누지 말고!
    global cnt_0
    global cnt_1
    if count_0(arr)==len(arr)**2:
        cnt_0+=1
    elif count_0(arr)==0:
        cnt_1+=1
    else:
        lab=len(arr)//2 ## lab*lab이 각 partition별 원소 갯수
        a=[] # 좌상
        for i in range(lab):
            a.append(arr[i][:lab])
        b=[] # 우상
        for i in range(lab):
            b.append(arr[i][lab:])
        c=[] # 좌하
        for i in range(lab,len(arr)):
            c.append(arr[i][:lab])
        d=[] # 우하
        for i in range(lab,len(arr)):
            d.append(arr[i][lab:])

        if count_0(a)==lab*lab:
            cnt_0+=1
        elif count_0(a)==0:
            cnt_1+=1
        else:
            DFS(a)
        #
        if count_0(b)==lab*lab:
            cnt_0+=1
        elif count_0(b)==0:
            cnt_1+=1
        else:
            DFS(b)
        #
        if count_0(c)==lab*lab:
            cnt_0+=1
        elif count_0(c)==0:
            cnt_1+=1
        else:
            DFS(c)
        #
        if count_0(d)==lab*lab:
            cnt_0+=1
        elif count_0(d)==0:
            cnt_1+=1
        else:
            DFS(d)

        
cnt_0=cnt_1=0
def solution(arr):
    global cnt_0
    global cnt_1
    DFS(arr)
    return [cnt_0,cnt_1]
```

## 방금 그 곡

```python
def change_code(code): ## A# ~ G#을 각각 1~5로 바꿔주는 함수
    new_code=''
    for idx,x in enumerate(code):
        if x=='#':
            if code[idx-1]=='A': # A#
                new_code=new_code[:len(new_code)-1]+'1'
            elif code[idx-1]=='C': # C#
                new_code=new_code[:len(new_code)-1]+'2'
            elif code[idx-1]=='D': # D#
                new_code=new_code[:len(new_code)-1]+'3'
            elif code[idx-1]=='F': # F#
                new_code=new_code[:len(new_code)-1]+'4'
            else: # G#
                new_code=new_code[:len(new_code)-1]+'5'
        else:
            new_code+=x            
    return new_code
            
def separate(musicinfos): 
    a=[]
    for x in musicinfos:
        aa=[]
        b=x.split(',')
        aa.append((int(b[1].split(':')[0])-int(b[0].split(':')[0]))*60+int(b[1].split(':')[1])-int(b[0].split(':')[1])) # 재생시간
        aa.append(b[2]) # 곡 제목
        aa.append(change_code(b[-1])) # 코드
        a.append(aa)
    ## a의 각 원소에는 재생시간,곡 제목, 바뀐 코드가 순서대로 있게 됨
    a.sort(key=lambda x: -x[0]) ## 이거 .sort(reverse=True)랑 같은 뜻 아녔음? 근데 reverse=True에선 순서 유지 안 되네
    b=[]
    for run_time,name,code in a:
        b.append([name,code*divmod(run_time,len(code))[0]+code[:divmod(run_time,len(code))[1]]])
    return b  ## 곡 제목, 재생시간까지 고려한 코드
                

def solution(m, musicinfos):
    m=change_code(m)
    music=separate(musicinfos)
    for name,code in music:
        if m in code:
            return name
    return '(None)'
```

```python
def solution(s):
    if s.count('(')==len(s) or s.count(')')==len(s):
        return False
    else:
        stack=[]
        for x in s:
            if x=='(':
                stack.append(x)
            else:
                if stack:
                    stack.pop()
                else:
                    return False
        if stack:
            return False
        else:
            return True
```

## 파일명 정렬

```python
def separate(file):
    file+=' ' ## 이거 없었을 때는 25점이었음. 근데 파일명이 숫자로 끝나는 경우도 있으니까 이거 붙여줘봤는데 바로 100점됨
    i=0
    while True:
        if file[i].isdigit():
            break # 첫 번째 digit index
        i+=1
    j=i
    while True:
        if not file[j].isdigit():
            break # 첫 번쨰 digit index 후, 첫 번째 string index
        j+=1
    return [file[:i].lower(),file[i:j],file[:len(file)-1]] ## head,number,file name 순

def solution(files):
    b=[separate(x) for x in files]
    b.sort(key=lambda x: (x[0],int(x[1])))
    return [x[-1] for x in b]
```

## n진수 게임

```python
def nzin(x,n):
    def DFS(x,n):
        nonlocal a
        if x==0:
            return # 그냥 return은 함수를 종료
        else:
            DFS(x//n,n)
            if x%n<10:
                a+='{}'.format(x%n)
            elif x%n==10:
                a+='A'
            elif x%n==11:
                a+='B'
            elif x%n==12:
                a+='C'
            elif x%n==13:
                a+='D'
            elif x%n==14:
                a+='E'
            else:
                a+='F'               
    a=''
    DFS(x,n)
    return a

def solution(n,t,m,p):
    b='0'
    for i in range(1,m*t+1):
        b+=nzin(i,n)
        
    s=b[p-1]
    z=p-1+m
    while True:
        if len(s)==t:
            break
        s+=b[z]
        z+=m
    
    return s
```
