# 가장 큰 정사각형 찾기



```python
def solution(board):
    answer = 0
    if len(board) == 1 or len(board[0]) == 1:
        if board[0][0] == 0:
            return 0
        else:
            return 1
    for i in range(1,len(board)):
        for j in range(1,len(board[0])):
            if board[i][j] != 0:
                board[i][j] = min(board[i-1][j],board[i][j-1],board[i-1][j-1]) + 1
            answer = max(answer, board[i][j])
    return answer**2

# 이 문제가 왜 DP일까?
# -> https://velog.io/@ju_h2/Python-%ED%94%84%EB%A1%9C%EA%B7%B8%EB%9E%98%EB%A8%B8%EC%8A%A4-level2-%EA%B0%80%EC%9E%A5-%ED%81%B0-%EC%A0%95%EC%82%AC%EA%B0%81%ED%98%95-%EC%B0%BE%EA%B8%B0-%EB%8F%99%EC%A0%81-%ED%94%84%EB%A1%9C%EA%B7%B8%EB%9E%98%EB%B0%8D-dp
# 정사각형 변의 길이로 접근한다는 점... 유레카
```

```python
import re
def convert(matchObj):
        string = matchObj.group()[0]
        return string.lower()
def solution(m, musicinfos):
    answer = ''
    m = re.sub(pattern = '\w#', repl = convert, string = m)
    info = []
    for musicinfo in musicinfos:
        start, end, title, note = re.split(',',musicinfo)
        s_h, s_m = list(map(int,re.split(':',start)))
        e_h, e_m = list(map(int,re.split(':',end)))
        length = (e_h-s_h)*60 + (e_m-s_m)
        note = re.sub(pattern = '\w#', repl = convert, string = note)
        i = 0
        if len(note) > length:
            note = note[:length]
        else:
            while len(note)<length:
                note += note[i]
                i += 1
        if m in note:
            info.append([length, title, note])

    if not info:
        return "(None)"
    else:
        info.sort(key = lambda x: x[0], reverse = True)
        return info[0][1]

# 처음에 문제 읽을땐 난해했지만 풀다보니 어떻게 풀리긴 하더라
# C# 같은 부분은 그냥 소문자 c로 변환했더니 생각하기가 쉬워졌음
# 배운점
# 1. re를 써서 치환할때 replace에는 함수가 들어갈수 있음

# 좋아요 많은 풀이
def ttt(ts, te):
    tmp = (int(te.split(":")[0]) - int(ts.split(":")[0])) * 60 + int(te.split(":")[1]) - int(ts.split(":")[1])
    return tmp
def con(s):
    while s.find("#") != -1:
        a = s.find("#")
        s = s[:a - 1] + s[a - 1].lower() + s[a + 1:]
    return s
def solution(m, musicinfos):
    m = con(m)
    ans = ''
    temp = 0
    for i in musicinfos:
        mu = i.split(',')[2]
        ss = i.split(',')[3]
        ti = ttt(i.split(',')[0], i.split(',')[1])
        ss = con(ss)
        tttt = ''
        ttmp = ti
        while ttmp > ss.__len__():
            tttt += ss
            ttmp -= ss.__len__()
        tttt += ss[:ti]
        if m in tttt and temp < ti:
            ans = mu
            temp = ti
    return ans if ans != '' else "(None)"

# 정규표현식 안쓰고도 간단하게 나눠질수가 있다보니 이렇게 한듯?

def shap_to_lower(s):
    s = s.replace('C#','c').replace('D#','d').replace('F#','f').replace('G#','g').replace('A#','a')
    return s

def solution(m,musicinfos):
    answer=[0,'(None)']   # time_len, title
    m = shap_to_lower(m)
    for info in musicinfos:
        split_info = info.split(',')
        time_length = (int(split_info[1][:2])-int(split_info[0][:2]))*60+int(split_info[1][-2:])-int(split_info[0][-2:])
        title = split_info[2]
        part_notes = shap_to_lower(split_info[-1])
        full_notes = part_notes*(time_length//len(part_notes))+part_notes[:time_length%len(part_notes)]
        if m in full_notes and time_length>answer[0]:
            answer=[time_length,title]
    return answer[-1]

# 내 코드를 축약하면 이런느낌일거 같다
```

# [3차] n진수 게임


```python
def solution(n, t, m, p):
    def function(number, n):
        L = "0123456789ABCDEF"
        q, r = divmod(number, n)
        if q == 0:
            return L[r]
        else:
            return function(q, n) + L[r]
    number = 0
    answer = ''
    while len(answer) < t*m:
        a = function(number, n)
        answer += a
        number += 1
    answer = [answer[p-1+i*m] for i in range(len(answer)//m)]
    return ''.join(answer)

# 재귀함수로 n진법 함수 만들고, 만들어야 하는 갯수 * 전체 명수로 다 뽑아서
# 해당 순서만큼만 출력하는 걸로

# 좋아요 가장 많은 풀이
big = ["A","B","C","D","E","F"]
def solution(n, t, m, p):
    a="0"
    i=0
    while True:
        if len(a)>=t*m:
            break
        b=""
        j=i
        while (j):
            if j%n>9:
                b=big[j%n-10]+b
            else:
                b=str(j%n)+b
            j=j//n
        a=a+b
        i=i+1
    answer = a[p-1::m][:t]
    return answer
```
