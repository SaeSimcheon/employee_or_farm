# 다단계 칫솔 판매

```python
from collections import defaultdict
def solution(enroll, referral, seller, amount):
    answer = {}
    for i in enroll:
        answer[i] = 0
    amount = [i * 100 for i in amount]
    multi_list = defaultdict(list)
    for a, b in zip(enroll, referral):
        multi_list[a].append(b)
    idx = 0
    while idx < len(amount):
        seller_idx, price_idx = seller[idx], amount[idx]
        answer[seller_idx] += price_idx
        while True:
            if price_idx//10 >= 1:
                price_idx = price_idx//10
                answer[seller_idx] -= price_idx
                seller_idx = multi_list[seller_idx][0]
                if seller_idx == "-":
                    break
                else:
                    answer[seller_idx] += price_idx
            else:
                break
        idx += 1
    return list(answer.values())

# 단순 코드 구현 문제인거 같다

# 좋아요 가장 많은 풀이
def solution(enroll, referral, seller, amount):
    answer = [0]*len(enroll)
    idx_list={}
    for idx,name in enumerate(enroll):
        idx_list[name]=idx
    for idx,name in enumerate(seller):
        price=100*amount[idx]
        answer[idx_list[name]]+=price
        while referral[idx_list[name]]!="-":
            answer[idx_list[name]]-=price//10
            name=referral[idx_list[name]]
            answer[idx_list[name]]+=price//10
            price=price//10
            if price==0:
                break
        answer[idx_list[name]]-=price//10
    return answer
```

# 경주로 건설

```python
def solution(board):
    dx = [0, -1, 1, 0]
    dy = [-1, 0, 0, 1]
    direction = ['U', 'L', 'R', 'D']
    size = len(board)
    board[0][0] = 0
    stack = [[0, 0, 'start']]
    while stack:
        x, y, current = stack.pop()
        for i in range(4):
            xx, yy, next_dir = x + dx[i], y + dy[i], direction[i]
            cost = 0
            if 0 <= xx < size and 0 <= yy < size and board[xx][yy] != 1:
                cost += 100
                if current != next_dir:
                    cost += 500
                if board[xx][yy] == 0 or board[xx][yy] >= board[x][y] + cost:
                    board[xx][yy] = board[x][y] + cost
                    stack.append([xx, yy, next_dir])
    return board[-1][-1]-500

# 이동한 과정을 단순히 저장하면 무조건 틀리더라
# 방향 설정을 어떻게 하느냐에 따라서 케이스 한개씩 틀리고 안틀리고... 왜 그러지
# 이동마다 최소 비용을 저장해야지 맞는듯..? dfs로도 해봤는데 왜 안되는지 모르겠음

```

# 등굣길

```python
def solution(m, n, puddles):
    board = [[1]*m for _ in range(n)]
    for x, y in puddles:
        board[y-1][x-1] = 0
        if y == 1:
            for yy in range(x-1,m):
                board[0][yy] = 0
        if x == 1:
            for xx in range(y-1,n):
                board[xx][0] = 0
    for i in range(1,n):
        for j in range(1,m):
            if board[i][j] != 0:
                board[i][j] = board[i-1][j] + board[i][j-1]
    return board[n-1][m-1]%1000000007

# 최소거리 구할때 bottom-up으로 하는거는 알았는데
# 웅덩이가 사이드에 있을경우도 고려했어야했음

# 좋아요 가장 많은 풀이
def solution(m,n,puddles):
    grid = [[0]*(m+1) for i in range(n+1)] #왼쪽, 위로 한줄씩 만들어서 IndexError 방지
    if puddles != [[]]:                    #물이 잠긴 지역이 0일 수 있음
        for a, b in puddles:
            grid[b][a] = -1                #미리 -1로 체크
    grid[1][1] = 1
    for j in range(1,n+1):
        for k in range(1,m+1):
            if j == k == 1:                #(1,1)은 1로 만들어두고, 0이 되지 않도록
                continue
            if grid[j][k] == -1:           #웅덩이는 0으로 만들어 다음 덧셈 때 영향끼치지 않게
                grid[j][k] = 0
                continue
            grid[j][k] = (grid[j][k-1] + grid[j-1][k])%1000000007   #[a,b] = [a-1,b] + [a,b-1] 공식
    return grid[n][m]


def solution(m, n, puddles):
    answer = 0
    info = dict([((2, 1), 1), ((1, 2), 1)])
    for puddle in puddles:
        info[tuple(puddle)] = 0
    def func(m, n):
        if m < 1 or n < 1:
            return 0
        if (m, n) in info:
            return info[(m, n)]
        return info.setdefault((m, n), func(m - 1, n) + func(m, n - 1))
    return  func(m, n) % 1000000007

# 재귀함수 풀이인데 테이블을 만드는게 아니라 지나간곳을 딕셔너리에 저장하는 형식
# 졸라 신기하네
```
